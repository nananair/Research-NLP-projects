```python
import pandas as pd

df = pd.read_excel('/content/drive/My Drive/updated_thesis_dataset.xlsx')
df.info()
pd.set_option("display.max_colwidth", 100)
df.head()
```

```python
# Chi-square tests of association between news values and share quartiles

from scipy.stats import chi2_contingency

news_values = [
    "Eliteness", "Personalisation", "Superlativeness", "Negativity",
    "Timeliness", "Unexpectedness", "Impact", "Space", "Positivity"
]

chi2_results = []

for case in df["Dataset"].unique():
    case_df = df[df["Dataset"] == case].copy()
    case_df["Share_quartile"] = pd.qcut(case_df["Share counts"], q=4, labels=["Q1", "Q2", "Q3", "Q4"])

    for nv in news_values:
        binary_var = "High_" + nv
        case_df[binary_var] = (case_df[nv] > case_df[nv].median()).astype(int)
        contingency = pd.crosstab(case_df[binary_var], case_df["Share_quartile"])
        if contingency.shape == (2, 4):
            chi2, p, dof, expected = chi2_contingency(contingency)
            chi2_results.append({
                "Case Study": case,
                "News Value": nv,
                "Chi2": chi2,
                "p-value": p,
                "Degrees of Freedom": dof
            })

results_df = pd.DataFrame(chi2_results)
print(results_df)

# results_df.to_excel('/content/drive/My Drive/chi_square_results_by_case.xlsx', index=False)
```

```python
# Average news value scores across share quartiles

import matplotlib.pyplot as plt
import seaborn as sns

avg_score_tables = []

for case in df["Dataset"].unique():
    case_df = df[df["Dataset"] == case].copy()
    case_df["Share Quartile"] = pd.qcut(case_df["Share counts"], q=4, labels=["Q1", "Q2", "Q3", "Q4"])
    grouped = case_df.groupby("Share Quartile")[news_values].mean().reset_index()
    grouped["Case Study"] = case
    grouped = grouped[["Case Study", "Share Quartile"] + news_values]

    avg_score_tables.append(grouped)

final_table = pd.concat(avg_score_tables, ignore_index=True)
final_table
```
